•	10 work groups each including 150 users in total<br>
•	Duration Range of Data: 6 weeks<br>
	User role Distribution:<br>
•	3% Admins<br>
•	10% Support Staff Shift Employees, at least 2 work groups<br>
•	30% Regular Shift Employees, at least 2 work groups<br>
•	53% Regular Day Employees, at least 2 work groups<br>
•	5% Tolerance in the number of employees distribution across roles

```python
import numpy as np
import pandas as pd

# Define constants
num_groups = 10
users_per_group = 150
total_users = num_groups * users_per_group
duration_weeks = 6
days_per_week = 7
timeframe_days = duration_weeks * days_per_week

# User role distributions
roles = ['Admin', 'Support Staff', 'Regular Shift', 'Regular Day']
role_distribution = {
    'Admin': 0.03,
    'Support Staff': 0.10,
    'Regular Shift': 0.30,
    'Regular Day': 0.53
}
role_tolerance = 0.05

# Ensure minimum groups for certain roles
min_groups_for_roles = {
    'Support Staff': 2,
    'Regular Shift': 2,
    'Regular Day': 2
}

# Generate user data
user_data = []
user_id = 0
for group_id in range(1, num_groups + 1):
    for _ in range(users_per_group):
        user_data.append([user_id, group_id])
        user_id += 1

user_df = pd.DataFrame(user_data, columns=['UserID', 'GroupID'])

# Assign roles to users ensuring distribution constraints
role_counts = {role: int(total_users * proportion) for role, proportion in role_distribution.items()}
remaining_users = total_users - sum(role_counts.values())

# Adjust for tolerance
for role in role_counts:
    role_counts[role] = int(role_counts[role] * (1 + np.random.uniform(-role_tolerance, role_tolerance)))

# Ensure the total count is still valid
role_counts['Regular Day'] += remaining_users  # Adjust last role to ensure total matches

# Assign roles ensuring minimum groups
user_roles = []
for role, count in role_counts.items():
    if role in min_groups_for_roles:
        groups = np.random.choice(num_groups, min_groups_for_roles[role], replace=False)
    else:
        groups = np.random.choice(num_groups, num_groups, replace=False)

    users_assigned = 0
    for group in groups:
        group_users = user_df[user_df['GroupID'] == group]['UserID'].values
        num_users = min(count - users_assigned, len(group_users))
        user_roles.extend([(user, role) for user in group_users[:num_users]])
        users_assigned += num_users
        if users_assigned >= count:
            break

# Assign remaining roles randomly
remaining_users = set(user_df['UserID'].values) - {user for user, _ in user_roles}
for user in remaining_users:
    role = np.random.choice(roles, p=[role_distribution[r] for r in roles])
    user_roles.append((user, role))

role_df = pd.DataFrame(user_roles, columns=['UserID', 'Role'])

# Merge user data with roles
user_df = user_df.merge(role_df, on='UserID')

# Generate time data
date_range = pd.date_range(start='2023-01-01', periods=timeframe_days, freq='D')
time_df = pd.DataFrame({'Date': date_range})

# Create a DataFrame with all possible user-date combinations
user_time_df = pd.MultiIndex.from_product([user_df['UserID'], time_df['Date']], names=['UserID', 'Date']).to_frame(index=False)

# Merge with user data to get full DataFrame
final_df = user_time_df.merge(user_df, on='UserID')

print(final_df.head())
```

Generating Skewed distribution with kurtesis.

```python
import scipy
import plotly.graph_objects as go
import numpy as np
from random import random

distribution = scipy.stats.gengamma(300, 0.1, loc=50, scale=1)
pre_samples = distribution.rvs(size=100000)  # Generate 1000 random samples
_min = pre_samples.min() - random()*2
_max = pre_samples.max() + random()*2
# Apply CDF transformation to map samples to the range [0, 24]
cdf_values = distribution.cdf(pre_samples)  # Map to [0, 1]
samples = cdf_values * 24  # Scale to [0, 24]
samples = (pre_samples - _min) / (_max - _min) * 24
# Create a histogram of the samples
histogram = go.Histogram(x=samples, nbinsx=200, name='Samples', opacity=0.75)
fig = go.Figure(data=[histogram])
fig.update_layout(
    title='Generalized Gamma Distribution', xaxis_title='Value', yaxis_title='Frequency', barmode='overlay'
)
# Show the plot
fig.show()
```

```python
import random

# Function to generate a random IP address within a given range
def generate_ip(base_ip, subnet_mask):
    # Extract base IP and convert to a list of integers
    base_ip_parts = list(map(int, base_ip.split('.')))

    # Calculate the number of host bits from the subnet mask
    host_bits = 32 - subnet_mask

    # Calculate the number of possible hosts
    num_hosts = 2**host_bits - 2  # Subtracting 2 for network and broadcast addresses

    # Generate a random host part within the available range
    host_part = random.randint(1, num_hosts)

    # Convert the host part to a 4-part IP address
    host_ip_parts = [host_part >> 16, (host_part >> 8) & 0xFF, host_part & 0xFF]

    # Combine base IP and host part to form the full IP address
    random_ip = base_ip_parts[:3] + [host_ip_parts[2]]

    return '.'.join(map(str, random_ip))

# Generate 30 random IP addresses in the range 192.168.10.0/24
random_ips = [generate_ip("192.168.10.0", 24) for _ in range(30)]
random_ips
```
